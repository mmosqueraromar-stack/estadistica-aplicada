La covarianza mide la variación conjunta de dos variables respecto a sus medias aritméticas. En este caso, la covarianza entre *price* y *review_scores_value* es negativa (-0.49), lo que indica que existe una ligera relación inversa entre ambas variables: a medida que aumenta el precio, la valoración tiende a disminuir.

Sin embargo, al tratarse de un valor cercano a 0, la relación es débil, por lo que no se puede afirmar que exista una dependencia clara entre ambas variables. Si la covarianza fuese exactamente 0, indicaría independencia entre ellas.

Además, la covarianza no está estandarizada, por lo que su valor depende de la escala de las variables, siendo más adecuado utilizar el coeficiente de correlación para analizar la intensidad de la relación.

```python
import pandas as pd
import zipfile

# Descomprimir (solo si no lo has hecho ya)
with zipfile.ZipFile("Datos prácticas de evaluación-20260505.zip", "r") as zip_ref:
    zip_ref.extractall("datos")

# Cargar datos
df = pd.read_csv("datos/listings.csv")

# Seleccionar variables y eliminar nulos
df = df[["review_scores_cleanliness", "review_scores_value"]].dropna()

# Medias
mean_x = df["review_scores_cleanliness"].mean()
mean_y = df["review_scores_value"].mean()

# Covarianza (la recalculamos porque la necesitamos)
cov = ((df["review_scores_cleanliness"] - mean_x) *
       (df["review_scores_value"] - mean_y)).sum() / (len(df) - 1)

# Desviaciones típicas
std_x = df["review_scores_cleanliness"].std()
std_y = df["review_scores_value"].std()

# Coeficiente de Pearson (fórmula)
pearson = cov / (std_x * std_y)

pearson
```

El coeficiente de correlación de Pearson mide la fuerza y el sentido de la relación lineal entre dos variables cuantitativas, tomando valores entre -1 y 1.

En este caso, el coeficiente obtenido es r = 0.745, lo que indica una relación positiva fuerte entre las variables *review_scores_cleanliness* y *review_scores_value*. Esto implica que, a medida que aumenta la puntuación de limpieza, también tiende a aumentar la valoración global del alojamiento.

Según la magnitud del coeficiente, la relación puede considerarse fuerte, ya que se encuentra por encima de 0.7. Por tanto, existe una asociación lineal clara entre ambas variables.

A diferencia de la covarianza, el coeficiente de Pearson está estandarizado, por lo que permite comparar la intensidad de la relación entre distintas variables independientemente de su escala de medida.
